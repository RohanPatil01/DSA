# How to create a Binary Tree in Python

```python
class node:
    def __init__(self,value):
        self.left=None
        self.right=None
        self.value=value
        
if __name__=="__main__":
    root=node("A")
    root.left=node("B")
    root.right=node("C")
    root.left.right=node("D")
    root.right.left=node("F")
    root.right.right=node("H")
```
