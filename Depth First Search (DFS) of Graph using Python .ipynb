# Implementation of Depth First Search (DFS) Algorithm using python

```python
#Code to create empty structure
from collections import defaultdict 
class Graph():
    def __init__(self):
        self.graph=defaultdict(list)
    def connectEdge(self,startNode,endNode):
        self.graph[startNode].append(endNode)
    def DFS_recur(self,node,visited):
        visited.add(node)
        print(node,end=" ")
        for node in self.graph[node]:
            if node not in visited:
                self.DFS_recur(node,visited)

    def DFS(self,startNode):
        visited=set()
        self.DFS_recur(startNode,visited)
```

```python
#Driver Code
if __name__=="__main__":
    root=Graph()
    root.connectEdge("A","C")
    root.connectEdge("A","B")
    root.connectEdge("A","F")
    root.connectEdge("C","A")
    root.connectEdge("C","D")
    root.connectEdge("C","F")
    root.connectEdge("B","A")
    root.connectEdge("B","D")
    root.connectEdge("B","E")
    root.connectEdge("B","F")
    root.connectEdge("D","C")
    root.connectEdge("D","B")
    root.connectEdge("D","E")
    root.connectEdge("E","B")
    root.connectEdge("E","D")
    root.connectEdge("F","A")
    root.connectEdge("F","C")
    root.connectEdge("F","B")    
```
