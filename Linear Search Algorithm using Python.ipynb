```python
def linearSearch(k,array):
    for element in array:
        if element==k:
            return array.index(element)
    else:
        return -1
```

```python
if __name__=="__main__":
    array=[12,9,35,10,1,22,7,57,17,2]
    searchElement=int(input("Enter the number that you want to search : "))
    result=linearSearch(searchElement,array)
    if result==-1:
        print("The element that you are looking for is not present in the given array.")
    else:
        print("The element is present at the {}th index".format(result))
```

```python
def search(arr, N, x):
    for i in range(0, N):
        if (arr[i] == x):
            return i
    return -1
if __name__ == "__main__":
    arr = [2, 3, 4, 10, 40]
    x = 10
    N = len(arr)
    result = search(arr, N, x)
    if(result == -1):
        print("Element is not present in array")
    else:
        print("Element is present at index", result)
```
